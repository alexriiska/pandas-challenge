```python
# District Summary
# Create a high-level snapshot, in a DataFrame, of the district's key metrics, including the following:

# Total schools
# Total students
# Total budget
# Average math score
# Average reading score
# % passing math (the percentage of students who passed math)
# % passing reading (the percentage of students who passed reading)
# % overall passing (the percentage of students who passed math AND reading)


import pandas as pd
from statistics import mean


#read the files as dataframes
schools_csv = "Resources/schools_complete.csv"
schools_df = pd.read_csv(schools_csv)

students_csv = "Resources/students_complete.csv"
students_df = pd.read_csv(students_csv)
```

```python
# Total schools
num_schools = schools_df['School ID'].count()
print(f'Total Schools in District: {num_schools}')
```

```python
# Total students
num_students = students_df['Student ID'].count()
print(f'Total Students in District: {num_students:,}')
```

```python
# Total budget
total_budget = schools_df['budget'].sum()
print(f'Total District Budget: ${total_budget:,}')
```

```python
# Average math score
avg_math = mean(students_df['math_score'])
print(f'District Average Math Score: {avg_math:.2f}')
```

```python
# Average reading score
avg_reading = mean(students_df['reading_score'])
print(f'District Average Reading Score: {avg_reading:.2f}')
```

```python
# % passing math (the percentage of students who passed math)
#asssumption: passing score: 70
math_pass = students_df.loc[students_df['math_score'] >= 70,:]
math_pass_count = (math_pass['Student ID'].count() / num_students)*100
print(f'Percent of Students in District that Passed Math: {math_pass_count:.2f}%')
```

```python
# % passing reading (the percentage of students who passed reading)
#asssumption: passing score: 70
reading_pass = students_df.loc[students_df['reading_score'] >= 70,:]
reading_pass_count = (reading_pass['Student ID'].count() / num_students)*100
print(f'Percent of Students in District that Passed Reading: {reading_pass_count:.2f}%')
```

```python
# % overall passing (the percentage of students who passed math AND reading)
overall_pass =  students_df.loc[(students_df['reading_score'] >= 70) & (students_df['math_score'] >= 70)]
overall_pass_count =(overall_pass['Student ID'].count() / num_students)*100
print(f'Percent of Students in District that Passed Math & Reading: {overall_pass_count:.2f}%')
```

```python
# Create a high-level snapshot, in a DataFrame, of the district's key metrics, including the following:
#Total schools
# Total students
# Total budget
# Average math score
# Average reading score
# % passing math (the percentage of students who passed math)
# % passing reading (the percentage of students who passed reading)
# % overall passing (the percentage of students who passed math AND reading)
district_df = pd.DataFrame({'Total Schools':num_schools,
                            'Total Students': num_students,
                            'Total Budget':[total_budget],
                            'Average Math Score':[avg_math],
                            'Average Reading Score':[avg_reading],
                            "% Passing Math":[math_pass_count],
                            "% Passing Reading":[reading_pass_count],
                            "% Overall Passing":[overall_pass_count]                         
                            })

#fomratting the columns
district_df['Total Budget'] = district_df['Total Budget'].map('${:,.2f}'.format)
district_df['Total Students'] = district_df['Total Students'].map('{:,}'.format)
district_df['Average Math Score'] = district_df['Average Math Score'].map('{:.2f}'.format)
district_df['Average Reading Score'] = district_df['Average Reading Score'].map('{:.2f}'.format)
district_df["% Passing Math"] = district_df["% Passing Math"].map('{:.2f}'.format)
district_df["% Passing Reading"] = district_df["% Passing Reading"].map('{:.2f}'.format)
district_df["% Overall Passing"] = district_df["% Overall Passing"].map('{:.2f}'.format)



district_df
```

```python
# School Summary

# School name
# School type
school_type = schools_df.set_index(['school_name'])['type']
school_type
```

```python
#merge the dataframes
df_merge = pd.merge(schools_df, students_df, on=['school_name'])
df_merge.head()

```

```python
# Total school budget

budget_per_school = schools_df.set_index(['school_name'])['budget']
# budget_per_school = budget_per_school.map('${:,}'.format)

budget_per_school
```

```python
# Per student budget

budget_per_student = budget_per_school / students_per_school
#budget_per_student = budget_per_student.map('${:.0f}'.format)
budget_per_student
```

```python
# Average math score

math_per_school = round(df_merge.groupby(['school_name']).mean()['math_score'],2)

math_per_school
```

```python
# Average reading score
reading_per_school = round(df_merge.groupby(['school_name']).mean()['reading_score'],2)

reading_per_school
```

```python
# % passing math (the percentage of students who passed math)


math_pass_per_school = df_merge.loc[df_merge['math_score'] >= 70,:]
math_pass_per_school_count = math_pass_per_school.groupby(['school_name']).count()['Student ID']
pct_math_pass_per_school = round((math_pass_per_school_count/students_per_school)*100,2)

pct_math_pass_per_school
```

```python
# % passing reading (the percentage of students who passed reading)

reading_pass_per_school = df_merge.loc[df_merge['reading_score'] >= 70,:]
reading_pass_per_school_count = reading_pass_per_school.groupby(['school_name']).count()['Student ID']
pct_reading_pass_per_school = round((reading_pass_per_school_count/students_per_school)*100,2)

pct_reading_pass_per_school
```

```python
# % overall passing (the percentage of students who passed math AND reading)

overall_pass_per_school =  df_merge.loc[(df_merge['reading_score'] >= 70) & (df_merge['math_score'] >= 70)]
overall_pass_per_school_count = overall_pass_per_school.groupby(['school_name']).count()['Student ID']
pct_overall_pass_per_school = round((overall_pass_per_school_count/students_per_school)*100,2)

pct_overall_pass_per_school
```

```python
# Create a DataFrame that summarizes key metrics about each school

school_summary_df = pd.DataFrame({'School Type':school_type,
                                  'Total Students':students_per_school,
                                  'Total School Budget':budget_per_school,
                                  'Per Student Budget':budget_per_student,
                                  'Average Math Score':math_per_school,
                                  'Average Reading Score':reading_per_school,
                                  "% Passing Math":pct_math_pass_per_school,
                                  "% Passing Reading":pct_reading_pass_per_school,
                                  "% Overall Passing":pct_overall_pass_per_school
                                 })

#formatting

school_summary_df['Total School Budget'] = school_summary_df['Total School Budget'].map('${:,.2f}'.format)
school_summary_df['Per Student Budget'] = school_summary_df['Per Student Budget'].map('${:,.2f}'.format)
school_summary_df['Total Students'] = school_summary_df['Total Students'].map('{:,}'.format)
school_summary_df['Average Math Score'] = school_summary_df['Average Math Score'].map('{:.2f}'.format)
school_summary_df['Average Reading Score'] = school_summary_df['Average Reading Score'].map('{:.2f}'.format)
school_summary_df["% Passing Math"] = school_summary_df["% Passing Math"].map('{:.2f}'.format)
school_summary_df["% Passing Reading"] = school_summary_df["% Passing Reading"].map('{:.2f}'.format)
school_summary_df["% Overall Passing"] = school_summary_df["% Overall Passing"].map('{:.2f}'.format)

school_summary_df
                                  
```

```python
# Create a new column called "School Sizes", and fill it with the values collected the slicing.
#the argument on the left side of the equals sign is creating a new column name
# right side is using the .cut() function to create the new column name based on an existing column name

school_summary_df['School Size'] = pd.cut(school_summary_df['Total Students'], size_bins, labels=size_groups, include_lowest=True)

#print the results to check the groups are correct
school_summary_df
```
